```python
import pandas as pd
df = pd.read_csv('./student-scores.csv')
df.head()
inputs = df.drop('career_aspiration', axis='columns')
targets = df['career_aspiration']
targets
from sklearn.preprocessing import LabelEncoder
le_id_ = LabelEncoder()
le_first_name = LabelEncoder()
le_last_name = LabelEncoder()
le_email = LabelEncoder()
le_gender = LabelEncoder()
le_part_time_job = LabelEncoder()
le_absence_days = LabelEncoder()
le_extracurricular_activities = LabelEncoder()
le_weekly_self_study_hours = LabelEncoder()
le_math_score = LabelEncoder()
le_history_score = LabelEncoder()
le_physics_score = LabelEncoder()
le_chemistry_score = LabelEncoder()
le_biology_score = LabelEncoder()
le_english_score = LabelEncoder()
le_geography_score = LabelEncoder()

inputs['id'] = le_id_.fit_transform(inputs['id'])
inputs['first_name'] = le_first_name.fit_transform(inputs['first_name'])
inputs['last_name'] = le_last_name.fit_transform(inputs['last_name'])
inputs['email'] = le_email.fit_transform(inputs['email'])
inputs['gender'] = le_gender.fit_transform(inputs['gender'])   
inputs['part_time_job'] = le_part_time_job.fit_transform(inputs['part_time_job'])
inputs['absence_days'] = le_absence_days.fit_transform(inputs['absence_days'])
inputs['extracurricular_activities'] = le_extracurricular_activities.fit_transform(inputs['extracurricular_activities'])
inputs['weekly_self_study_hours'] = le_weekly_self_study_hours.fit_transform(inputs['weekly_self_study_hours'])
inputs['math_score'] = le_math_score.fit_transform(inputs['math_score'])
inputs['history_score'] = le_history_score.fit_transform(inputs['history_score'])
inputs['physics_score'] = le_physics_score.fit_transform(inputs['physics_score'])
inputs['chemistry_score'] = le_chemistry_score.fit_transform(inputs['chemistry_score'])
inputs['biology_score'] = le_biology_score.fit_transform(inputs['biology_score'])
inputs['english_score'] = le_english_score.fit_transform(inputs['english_score'])
inputs['geography_score'] = le_geography_score.fit_transform(inputs['geography_score'])
inputs.head
# inputs_n = inputs.drop(['id', 'first_name', 'last_name', 'email', 'gender', 'part_time_job', 'absence_days', 'extracurricular_activities', 'weekly_self_study_hours', 'math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score'], axis='columns')
# inputs_n.head()

```
